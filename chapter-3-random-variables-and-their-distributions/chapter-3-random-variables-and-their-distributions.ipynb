# Distributions in R


```r
# Binomial Distribution

# Example: Bin(5,0.2)

dbinom(3,5,0.2) # PMF P(X = 3)
pbinom(3,5,0.2) # CDF P(X <= 3)
rbinom(7,5,0.2) # Generate 7 random Bin(5,0.2) variables
```

```r
# we can also evaluate a vector of values

dbinom(0:5,5,0.2)
pbinom(0:5,5,0.2)
```

```r
plot(pbinom(0:5,5,0.2),type = 'l')
```

```r
# Hypergeometric Distribution

# Example: HGeom(10,30,15)

dhyper(2,10,30,15) # PMF P(X = 2)
phyper(2,10,30,15) # CDF P(X <= 2)
rhyper(7,10,30,15) # Generate 7 random HGeom(10,30,15)
```

```r
# Discrete Distributions with finite support

x <- c(0,5,10,15)               # Supports
p <- c(0.25, 0.5, 0.1, 0.15)    # Corresponding Probability

sample(x, 100, prob=p, replace = TRUE)
```

```r
hist(sample(x, 100, prob=p, replace = TRUE))
```
